```python
# Find the mean sulphate present in different water samples collected.
np.mean(data["Sulfate"])
```

```python
# Fill the features having NULL with the mean value.
data.isna().sum()
```

```python
data["Sulfate"].fillna(np.mean(data["Sulfate"]),inplace=True)
```

```python
data["ph"].fillna(np.mean(data["ph"]),inplace=True)
```

```python
data["Trihalomethanes"].fillna(np.mean(data["Trihalomethanes"]),inplace=True)
```

```python
# Display the water samples where the ph is < 4 and chloramin is 5
df = data[(data["ph"] < 4) & (data["Chloramines"] < 5 )]
df.head(10)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df["ph"] = 4
```

```python
df
```

```python
# Convert the target column values as Drinkable where value is 1 and non- drinkable where value is 0.
data.Potability.replace({0:"Non-drinkable",1:"Drinkable"},inplace=True)
```

```python
data
```

```python
X = data.drop(["Potability"],axis=1)
y = data["Potability"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.15,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
SVC_model = SVC()
KNN_model = KNeighborsClassifier(n_neighbors=8)
```

```python
SVC_model.fit(X_train,y_train)
KNN_model.fit(X_train,y_train)
```

```python
svc_result = SVC_model.predict(X_test)
knn_result = KNN_model.predict(X_test)
```

```python
print(svc_result)
print(knn_result)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(svc_result,y_test))
```

```python
print(accuracy_score(knn_result,y_test))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,svc_result,labels=None))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,svc_result)
```

```python
confusion_matrix(y_test,knn_result)
```
