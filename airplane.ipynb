```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(["Class"],axis=1)
y = data["Class"]
```

```python
X["Gender"].unique()
```

```python
X.Gender.replace({"Male":0,"Female":1},inplace=True)
```

```python
X["Customer Type"].replace({"Loyal Customer":0,"disloyal Customer":1},inplace=True)
```

```python
X["Type of Travel"].unique()
```

```python
X["Type of Travel"].replace({"Personal travel":0,"Business travel":1},inplace=True)
```

```python
X.drop(["satisfaction"],axis=1,inplace=True)
```

```python
model = DecisionTreeClassifier(random_state=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.15,random_state=1)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
